#Praktikum 01: Indexing dan Slicing

```python
teks = "Politeknik Negeri Semarang"
print(teks[0])      #'P'
print(teks[0:11])   #'Politeknik'
print(teks[-8:])    #'Semarang'
```

#Praktikum 02: Method Bawaan String

```python
teks = "Politeknik Negeri Semarang"
print(teks.upper())
print(teks.lower())
print(teks.title())
print(teks.strip())
print(teks.split())
print("-".join(["Politeknik","Negeri","Semarang"]))
```

#Praktikum 05: Operasi dalam F-String

```python
a = int(input("Masukkan nilai a: "))
b = int(input("Masukkan nilai b: "))

print(f"Hasil penjumlahan {a}+{b} = {a+b}")
print(f"Hasil pembagian {a}/{b} = {a/b:.2f}")
```

#Praktikum 07: Studi Kasus Nyata: Analisis Kalimat

```python
kalimat = "Saya kuliah di Politeknik Negeri Semarang"
daftar_kata = kalimat.split(" ")

print(f"Jumlah kata: {len(daftar_kata)}")
for i, kata in enumerate (daftar_kata, start=1):
  print(i, kata)
```

#Tugas No 07: Operasi dalam F-String

```python
bilangan1 = int(input("Masukkan bilangan pertama: "))
bilangan2 = int(input("Masukkan bilangan kedua: "))

print(f"Hasil Penjumlahan: {bilangan1 + bilangan2}")
print(f"Hasil Perkalian: {bilangan1 * bilangan2}")
print(f"Hasil Pembagian: {bilangan1 / bilangan2:.3f}")
```

#Tugas No 08: Casting Data

```python
angka_str = input("Masukkan angka: ")

angka_int = int(angka_str)
kuadrat = angka_int ** 2
print(f"Kuadrat dari {angka_str} adalah: {kuadrat}")

angka_float = float(angka_str)
print(f"Angka {angka_str} (sebagai float) dengan 2 angka di belakang koma: {angka_float:.2f}")
```

#Tugas No 09: Studi Kasus - Analisis Teks

```python
kalimat = input("Masukkan sebuah kalimat: ")

kata_kata = kalimat.split()
jumlah_kata = len(kata_kata)
print(f"Jumlah kata dalam kalimat: {jumlah_kata}")

kata_terpanjang = max(kata_kata, key=len)
print(f"Kata terpanjang: {kata_terpanjang}")

kata_kata_sorted = sorted(kata_kata)
print("Daftar kata dalam urutan abjad:")
for kata in kata_kata_sorted:
  print(kata)
```
